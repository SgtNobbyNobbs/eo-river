### RAT IN packages

Functional requirements:

1. Get land boundary (cleaned water mask) from Google Earth Engine
2. Land boundary to network (skeletonize, topology)
3. For every branch, generate curvilinear grid (br, ch -> m, n)
4. Generate bathymetry for every curvilinear grid point
5. Write results as curvilinear grid (CF), GeoJSON -> TIF
